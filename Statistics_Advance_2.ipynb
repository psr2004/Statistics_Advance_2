# **Theoretical Questions**


1. What is hypothesis testing in statistics?
* Hypothesis testing is a statistical method that uses sample data to evaluate a hypothesis about a population parameter, determining whether there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
* The null hypothesis (H₀) is a statement of no effect or no difference that serves as the default assumption, while the alternative hypothesis (H₁) represents the presence of an effect or difference that the test aims to support.

3. What is the significance level in hypothesis testing, and why is it important?
* The significance level (α) is the probability threshold for rejecting the null hypothesis, controlling the risk of a Type I error; common choices like α = 0.05 balance sensitivity and error risk.

4. What does a P-value represent in hypothesis testing?
* The P-value is the probability of observing a test statistic at least as extreme as the one obtained, assuming the null hypothesis is true, quantifying the strength of evidence against H₀.

5. How do you interpret the P-value in hypothesis testing?
* A small P-value (≤ α) indicates strong evidence against H₀, leading to its rejection, whereas a large P-value (> α) suggests insufficient evidence and failure to reject H₀.

6. What are Type I and Type II errors in hypothesis testing?
* A Type I error occurs when H₀ is wrongly rejected (false positive), and a Type II error happens when H₀ is wrongly retained despite being false (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
* A one-tailed test evaluates an effect in a specific direction (e.g., greater than), allocating the entire α to one tail, while a two-tailed test assesses deviations in both directions, splitting α across both tails.

8. What is the Z-test, and when is it used in hypothesis testing?
* The Z-test compares sample and population means when the population standard deviation is known or the sample size is large, using the standard normal distribution of the test statistic.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
* The Z-score is calculated as (X̄ – μ₀) / (σ / √n), representing the standardized distance between the sample mean and the null hypothesis mean in units of standard error.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
* The T-distribution is a family of distributions with heavier tails, used for hypothesis tests when the population standard deviation is unknown and sample sizes are small, adjusting for extra uncertainty.

11. What is the difference between a Z-test and a T-test?
* The Z-test uses the known population standard deviation and a normal distribution, while the T-test uses the sample standard deviation and a T-distribution to account for smaller samples and unknown variance.

12. What is the T-test, and how is it used in hypothesis testing?
* The T-test assesses whether the mean of a sample differs from a known value (one-sample), or whether the means of two samples differ (independent or paired), using the T-distribution for inference.

13. What is the relationship between Z-test and T-test in hypothesis testing?
* Both tests compare means using standardized statistics; as sample size grows and variance estimates stabilize, the T-distribution converges to the normal, making T-tests approximate Z-tests.

14. What is a confidence interval, and how is it used to interpret statistical results?
* A confidence interval is a range of values, calculated from sample data, that is expected to contain the true population parameter with a specified probability (e.g., 95%), providing both estimate and uncertainty.

15. What is the margin of error, and how does it affect the confidence interval?
* The margin of error is the product of the critical value (Z or T) and the standard error, determining the half-width of the confidence interval; larger samples or lower confidence levels reduce it.

16. How is Bayes’ Theorem used in statistics, and what is its significance?
* Bayes’ Theorem updates the probability of a hypothesis based on new evidence by combining prior probability and likelihood, forming the basis for Bayesian inference and decision-making.

17. What is the Chi-square distribution, and when is it used?
* The Chi-square distribution arises from the sum of squared standard normal variables and is used in tests of variance, goodness of fit, and contingency tables.

18. What is the Chi-square goodness of fit test, and how is it applied?
* The Chi-square goodness of fit test compares observed category frequencies to expected frequencies under a hypothesized distribution, using the Chi-square statistic to assess fit quality.

19. What is the F-distribution, and when is it used in hypothesis testing?
* The F-distribution is the ratio of two scaled Chi-square distributions, used in ANOVA and tests comparing variances to determine if group variances or model fits differ.

20. What is an ANOVA test, and what are its assumptions?
* ANOVA (Analysis of Variance) tests for differences among multiple group means by partitioning total variance; it assumes independent samples, normally distributed errors, and equal variances (homoscedasticity).

21. What are the different types of ANOVA tests?
* Common types include one-way ANOVA (one factor), two-way ANOVA (two factors), and repeated measures ANOVA (same subjects measured under different conditions).

22. What is the F-test, and how does it relate to hypothesis testing?
* The F-test assesses whether group variances or model fits differ significantly by comparing variance estimates; in ANOVA, it tests whether at least one group mean differs from the others using the F-distribution.

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 100)
z_stat = (np.mean(data) - 50) / (10 / np.sqrt(len(data)))
x = np.linspace(-4, 4, 200)
pdf = stats.norm.pdf(x)
plt.plot(x, pdf)
plt.fill_between(x, pdf, where=(x <= -abs(z_stat)) | (x >= abs(z_stat)), color='red', alpha=0.3)
plt.title("Two-tailed Z-test Decision Region")
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type I and Type II errors during hypothesis testing
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def plot_errors(alpha):
    x = np.linspace(-4, 4, 400)
    pdf0 = stats.norm.pdf(x)
    shift = stats.norm.ppf(1 - alpha)
    plt.plot(x, pdf0, label='H₀ distribution')
    plt.fill_between(x, pdf0, where=(x >= shift), color='red', alpha=0.3, label='Type I error')
    pdf1 = stats.norm.pdf(x - shift)
    plt.fill_between(x, pdf1, where=(x <= shift), color='blue', alpha=0.3, label='Type II error')
    plt.legend()
    plt.show()

plot_errors(0.05)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    m = np.mean(data)
    se = stats.sem(data)
    h = se * stats.t.ppf((1 + confidence) / 2, n - 1)
    return m - h, m + h

ci = confidence_interval(np.random.randn(30))
print("Confidence interval:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

data = np.random.randn(40)
ci = confidence_interval(data, confidence=0.95)
margin_error = (ci[1] - ci[0]) / 2
print("Margin of error:", margin_error)
```

```python
# 11. Implement a Bayesian inference method using Bayes’ Theorem in Python and explain the process
prior = 0.5
likelihood = 0.7
evidence = prior * likelihood + (1 - prior) * 0.2
posterior = (likelihood * prior) / evidence
print("Posterior probability:", posterior)
```
